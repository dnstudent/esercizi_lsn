```python
import os
import numpy as np
import subprocess as proc
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2

executable_path = os.path.join("..", "cmake-build-release", "01")
seeds_path = os.path.join("..", "ariel_rng")
seeds_args = ["--primes="+os.path.join(seeds_path, "Primes"), "--seeds=" + os.path.join(seeds_path, "seed.in")]
data_root = os.path.join(".", "simulations")

def data_path(exp):
    return os.path.join(data_root, f"simulation_{exp}.csv")
```

```python
ex = "01_1"
data_1 = data_path(ex)
M = int(1E6)
N = int(1E2)
proc.run([os.path.join(executable_path, ex),
          f"--sample_size={M}", f"--n_blocks={N}", f"-o={data_1}"] + seeds_args)
data = pd.read_csv(data_1)
data.plot.line("block", "mean_estimate", yerr="mean_error", grid = True)
```

```python
data.variance_estimate -= 1/12
data.plot.line("block", "variance_estimate", yerr="variance_error", grid = True)
```

```python
ex = "01_1_3"
data_3 = data_path(ex)
M = int(1E6)
N = int(1E4)
R = int(1E2)
proc.run([os.path.join(executable_path, ex),
          f"--n_samples={N}", f"--sample_size={M}", f"--n_intervals={R}", f"-o={data_3}"] + seeds_args)

data = pd.read_csv(data_3)
data.plot("x", "chi_squared", grid = True)
```

```python
data["chi_squared"].plot.hist(density = True)
xs = np.linspace(min(data.chi_squared), max(data.chi_squared), 100)
plt.plot(xs, chi2(N).pdf(xs))
```
